```python
import pandas as pd
import numpy as np
```

```python
nomColmn = ['Username', 'Nb_Followers']
figaro = pd.read_csv('usernames_nbfollowers_figaro.csv', names=nomColmn)
figaro = figaro.dropna()
nomColmn = ['Username', 'Nb_Followers']
libe = pd.read_csv('usernames_nbfollowers_libe.csv', names=nomColmn)
libe = libe.dropna()
```

```python
print(figaro.shape)
print(libe.shape[0]+figaro.shape[0])
```

```python
figaro.head()
```

```python
nodes = pd.merge(figaro, libe, how='outer', on='Username')
print(nodes.shape)
nodes.head()
```

```python
hubs=[]

hubs.insert(0,{'Username':'Le_Figaro','Nb_Followers_x':np.nan, 'Nb_Followers_y':np.nan})
hubs.insert(0,{'Username':'libe','Nb_Followers_x':np.nan, 'Nb_Followers_y':np.nan })

nodes = pd.concat([pd.DataFrame(hubs), nodes], ignore_index=True)
nodes.head()
```

```python
nodes.index+=1
```

```python
nodes.head()
```

```python
nodes = nodes.iloc[0:10000]
print(nodes.shape)
nodes.head()
```

```python
base = nodes
base.head()
```

```python
nodes = nodes.Username
nodes.head()
```

```python
nodes.to_csv('nodes_10000.csv', index_label=['Id'] ,header=['Label'])
```

```python
def check_figaro(row):
    if(np.isnan(row['Nb_Followers_x'])):
        return np.nan
    else:
        return 2
    
def check_libe(row):
    if(np.isnan(row['Nb_Followers_y'])):
        return np.nan
    else:
        return 1
```

```python
base['Lien_Figaro']=base.apply(check_figaro, axis=1)
```

```python
base['Lien_Libe']=base.apply(check_libe, axis=1)
```

```python
base.tail()
```

```python
lien_figaro = base[['Lien_Figaro']]
lien_figaro.head()
```

```python
lien_figaro.reset_index(inplace=True)
lien_figaro.head()
```

```python
lien_libe = base[['Lien_Libe']]
lien_libe.head()
```

```python
lien_libe.reset_index(inplace=True)
lien_libe.head()
```

```python
lien_libe = lien_libe.dropna()
lien_libe.head(10)
```

```python
lien_figaro = lien_figaro.dropna()
lien_figaro.head(10)
```

```python
lien_figaro = lien_figaro.rename(columns={'index':'Source','Lien_Figaro':'Target'})
lien_figaro.head()
```

```python
lien_libe = lien_libe.rename(columns={'index':'Source','Lien_Libe':'Target'})
lien_libe.head()
```

```python
liens = pd.merge(lien_figaro, lien_libe, how='outer')
liens.head(10)
```

```python
liens[liens.Source==8].head()
```

```python
liens = liens.astype(int)
liens.head()
```

```python
liens.to_csv('liens_10000.csv',header=['Source','Target'], index=False,index_label=False)
```
